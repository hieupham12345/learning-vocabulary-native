```python
#insert Chinese
import sqlite3
import os

# ================= CẤU HÌNH =================
BASE_DIR = r"C:\Users\V3401226\Desktop\test\expo_app_training_language\vocab-app\data\HSK"
DB_PATH = r"C:\Users\V3401226\Desktop\test\expo_app_training_language\vocab-app\assets\vocab_words.db"
LANGUAGE = 'Chinese'
IS_LEARNED = 0

# Danh sách các file và level tương ứng
FILES_CONFIG = [
    {"file": "HSK_Level_1_words.txt", "level": "HSK1"},
    {"file": "HSK_Level_2_words.txt", "level": "HSK2"},
    {"file": "HSK_Level_3_words.txt", "level": "HSK3"},
    {"file": "HSK_Level_4_words.txt", "level": "HSK4"},
    {"file": "HSK_Level_5_words.txt", "level": "HSK5"},
    {"file": "HSK_Level_6_words.txt", "level": "HSK6"},
    {"file": "HSK_Level_7-9_words.txt", "level": "HSK7-9"},
]
# ============================================

def insert_all_hsk_levels():
    if not os.path.exists(os.path.dirname(DB_PATH)):
        os.makedirs(os.path.dirname(DB_PATH))

    conn = sqlite3.connect(DB_PATH)
    cursor = conn.cursor()

    # 1. Khởi tạo Database & Index
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS vocabulary (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            word TEXT NOT NULL,
            language TEXT NOT NULL,
            is_learned INTEGER NOT NULL DEFAULT 0,
            level TEXT
        )
    ''')
    cursor.execute('''
        CREATE UNIQUE INDEX IF NOT EXISTS idx_unique_word_lang_level 
        ON vocabulary(word, language, level)
    ''')

    total_added = 0

    for config in FILES_CONFIG:
        full_path = os.path.join(BASE_DIR, config["file"])
        level_name = config["level"]

        if not os.path.exists(full_path):
            print(f"⚠️ Bỏ qua: Không tìm thấy file '{config['file']}'")
            continue

        # Đọc dữ liệu từ file
        with open(full_path, 'r', encoding='utf-8') as file:
            words = [line.strip() for line in file.readlines() if line.strip()]

        if not words:
            print(f"⚠️ File '{config['file']}' trống.")
            continue

        # Chuẩn bị data cho level hiện tại
        data_to_insert = [(word, LANGUAGE, IS_LEARNED, level_name) for word in words]

        try:
            cursor.executemany('''
                INSERT OR IGNORE INTO vocabulary (word, language, is_learned, level)
                VALUES (?, ?, ?, ?)
            ''', data_to_insert)
            
            added_in_this_file = cursor.rowcount
            # Lưu ý: Trong một số phiên bản sqlite, rowcount sau executemany 
            # có thể không phản ánh chính xác số dòng inserted nếu dùng IGNORE.
            # Nhưng nó vẫn là cách tốt nhất để theo dõi sơ bộ.
            
            total_added += (added_in_this_file if added_in_this_file > 0 else 0)
            print(f"✅ Đã xử lý {level_name}: Thêm mới {added_in_this_file} từ.")

        except sqlite3.Error as e:
            print(f"❌ Lỗi DB tại {level_name}: {e}")

    conn.commit()
    conn.close()
    print(f"\n--- HOÀN TẤT TẤT CẢ ---")
    print(f"Dữ liệu đã được lưu tại: {DB_PATH}")

if __name__ == '__main__':
    insert_all_hsk_levels()
```

```python
#check db
import sqlite3

DB_PATH = r"C:\Users\V3401226\Desktop\test\expo_app_training_language\vocab-app\assets\vocab_words.db"

def check_database():
    conn = sqlite3.connect(DB_PATH)
    cursor = conn.cursor()

    try:
        # 1. Đếm tổng số từ
        cursor.execute("SELECT COUNT(*) FROM vocabulary")
        total = cursor.fetchone()[0]
        print(f"📊 Tổng số từ vựng hiện có: {total}")

        # 2. Thống kê theo Level
        print("\n📈 Thống kê theo Level:")
        cursor.execute("SELECT level, COUNT(*) FROM vocabulary GROUP BY level")
        levels = cursor.fetchall()
        for level, count in levels:
            print(f"   - {level}: {count} từ")

        # 3. Xem 5 từ đầu tiên
        print("\n🔍 Xem thử 5 từ đầu tiên:")
        cursor.execute("SELECT id, word, language, is_learned, level FROM vocabulary LIMIT 5")
        words = cursor.fetchall()
        for w in words:
            print(f"   {w}")

    except sqlite3.Error as e:
        print(f"❌ Có lỗi xảy ra: {e}")
    finally:
        conn.close()

if __name__ == '__main__':
    check_database()
```
